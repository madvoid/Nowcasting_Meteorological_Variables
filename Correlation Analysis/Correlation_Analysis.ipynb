# Correlation Analysis
Visualize correlation between input variables

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

## Load data

```python
col_names = ["LEMS I Wind U", 
             "LEMS J Wind U", 
             "LEMS K Wind U", 
             "LEMS I Wind V",
             "LEMS J Wind V",
             "LEMS K Wind V",
             "LEMS I Surface Temperature",
             "LEMS J Surface Temperature",
             "LEMS K Surface Temperature",
             "LEMS I Air Pressure",
             "LEMS J Air Pressure",
             "LEMS K Air Pressure",
             "LEMS I Specific Humidity",
             "LEMS J Specific Humidity",
             "LEMS K Specific Humidity",
            ]
df = pd.read_csv('Inputs.csv', names=col_names)
```

## Drop NaNs

```python
dfc = df.dropna(axis=0, how='any')
```

## Calculate Correlation Coefficients

```python
R = np.corrcoef(dfc, rowvar=False)
u_mask = np.logical_not(np.tril(np.ones(R.shape)))
```

## Plot Heatmap

```python
fig, ax = plt.subplots(1, 1, figsize=(17, 17))
sns.heatmap(data=R, vmin=-1.0, vmax=1.0, center=True, annot=True, square=True, xticklabels=col_names, yticklabels=col_names, mask=u_mask, ax=ax)
ax.set_title("Pearson's Correlation Coefficients for Model Inputs")
fig.savefig('Correlation_Heatmap.png', transparent=True)
```
